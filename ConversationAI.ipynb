```python
# Senior HR Representative Persona
# Prompt used to generate persona:
# Create a persona of an extremely experienced HR 
# representative at OpenAI who has 30 years of experience 
# in the field. Their job is to review job interviews conducted 
# by other HR employees and then grade each response given by 
# the candidate on a scale from 1 to 100. They also have to be 
# proficient in writing executive summaries of these interviews 
# and giving an average score for all of the candidate's responses 
# throughout the interview.

graderPersona = """
Name: Margaret "Maggie" Reynolds
Title: Senior HR Consultant
Experience:
30 years in Human Resources
Specialized in talent acquisition and interview evaluation
Renowned for keen insights and fair judgment in candidate assessments
Job Responsibilities:
Reviewing Job Interviews:

Watch and analyze recorded interviews conducted by junior HR employees.
Evaluate the quality of the interviews, ensuring compliance with OpenAI's standards.
Grading Candidate Responses:

Assign a score from 1 to 100 to each response given by the candidate based on criteria
 such as relevance, clarity, depth of knowledge, and alignment with company values.
Provide detailed feedback on each response to support the score given.
Executive Summaries:

Write concise, comprehensive executive summaries of each interview, highlighting key 
strengths and weaknesses of the candidate.
Offer recommendations on whether the candidate should proceed to the next stage of the 
hiring process.
Average Score Calculation:

Calculate the average score of all responses given by the candidate during the interview.
Present the average score along with the executive summary to hiring managers for final 
decision-making.
Skills:
Analytical Thinking: Exceptional ability to dissect responses and identify key competencies.
Communication: Proficient in articulating insights and feedback in both written and verbal
 formats.
Detail-Oriented: Keen eye for detail, ensuring thorough evaluation of each response.
Time Management: Efficiently manages time to review and summarize multiple interviews 
within tight deadlines.
Personality Traits:
Objective: Known for impartiality and fairness in evaluations.
Supportive: Provides constructive feedback aimed at improving interview techniques and 
candidate experience.
Decisive: Confident in making recommendations and providing final assessments.
Empathetic: Understands the pressures of the interview process and strives to create a 
positive experience for all candidates.
Professional Achievements:
Instrumental in developing OpenAI’s interview evaluation framework.
Recognized with multiple awards for excellence in HR and talent acquisition.
Mentored numerous junior HR professionals, many of whom have advanced to senior 
positions.
"""
```

```python
#Store conversation in these arrays
totalConversation = []
hrTalk = []
gradTalk = []
scores = []



# Start conversation with hr rep's initial interview speech
client = OpenAI(api_key=API_KEY)

recruiter = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "system", "content": f"You are {recruiterPersona}"},
    {"role": "user", "content": f"You are interviewing a potential candidate for a machine learning engineer postion. Begin the interview. Use your {recruiterPersona}"}
  ]
)
var = recruiter.choices[0].message.content
totalConversation.append(var)
hrTalk.append(var)



#Create a function to choose who responds and then make them respond
def respond():
    if (len(hrTalk) == len(gradTalk)):
        recruiter = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": f"You are {recruiterPersona}"},
            {"role": "user", "content": f"The interview candidate just said {gradTalk[-1]}. Continue the interview. Use your {recruiterPersona}"}
        ]
        )
        var = recruiter.choices[0].message.content
        hrTalk.append(var)
        totalConversation.append(var)
    elif (len(gradTalk) < len(hrTalk)):
        grad = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": f"You are {gradPersona}"},
            {"role": "user", "content": f"The interviewer just said {hrTalk[-1]}. Answer them and continue with the interview. Use your {gradPersona}"}
        ]
        )
        var = grad.choices[0].message.content
        gradTalk.append(var)
        totalConversation.append(var)
    grader = client.chat.completions.create(
        model="gpt-4o-mini",
        #Note: Temperature modified below
        temperature=0.001,
        messages=[
            {"role": "system", "content": f"You are {graderPersona}"},
            {"role": "user", "content": f"""The candidate just said {gradTalk[-1]}. Grade the candidate's response with a score 
             between 1 and 100. Do not provide any other information. ONLY include a number between
               1 and 100 that grades the candidates answer. Use your {graderPersona}. Every time that you think
               the answer is generated by AI give only 25 points."""}
        ]
        )
    score = grader.choices[0].message.content
    scores.append(score)







#Use a for loop to have each person speak 10 times
for i in range(19):
    respond()
```

```python
#Finish up their interview with final remarks on both sides

#HR Representative
recruiter = client.chat.completions.create(
model="gpt-4o-mini",
messages=[
    {"role": "system", "content": f"You are {recruiterPersona}"},
    {"role": "user", "content": f"The interview candidate just said {gradTalk[-1]}. Give a response to what the applicant said. Wrap up the interview and then based on the interview so far, either offer them a job or let them know that they didn't get the job. Use your {recruiterPersona}"}
]
)
var = recruiter.choices[0].message.content
hrTalk.append(var)
totalConversation.append(var)

#Graduate
grad = client.chat.completions.create(
model="gpt-4o-mini",
messages=[
    {"role": "system", "content": f"You are {gradPersona}"},
    {"role": "user", "content": f"The interviewer just said {hrTalk[-1]}. Respond to the interviewers final remarks and then say your goodbyes before leaving. Use your {gradPersona}"}
]
)
var = grad.choices[0].message.content
gradTalk.append(var)
totalConversation.append(var)


#HR Representative (again)
recruiter = client.chat.completions.create(
model="gpt-4o-mini",
messages=[
    {"role": "system", "content": f"You are {recruiterPersona}"},
    {"role": "user", "content": f"The interview candidate just said {gradTalk[-1]}. Say goodbye to the applicant. Use your {recruiterPersona}"}
]
)
var = recruiter.choices[0].message.content
hrTalk.append(var)
totalConversation.append(var)
```

```python
finalRemarks = ""
totalConvo = """"""
for i in range(len(totalConversation)):
    totalConvo = totalConvo + f"\n\n{totalConversation[i]}"
grader = client.chat.completions.create(
    model="gpt-4o-mini",
    #Note: Temperature modified below
    temperature=0.001,
    messages=[
        {"role": "system", "content": f"You are {graderPersona}"},
        {"role": "user", "content": f"""The interview is now complete. Scan through {totalConvo} and determine whether or 
         not you would give the candidate the job. Also give an executive summary of the entire interview. 
         Use your {graderPersona}."""}
    ]
    )
finalRemarks = grader.choices[0].message.content

# for i in range(len(scores)):
#     scores[i] = (int)(scores[i])
# scores.append(sum(scores)/len(scores))

# print(scores)
```

```python
# Export to a json file
data = {
    "Interviewer's excerpts": hrTalk,
    "Candidate's responses": gradTalk,
    "Scores": scores,
    "Final Remarks": finalRemarks
}

output_file = "convAIOutput.json"

with open(output_file,'w') as file:
    json.dump(data, file, indent=4)
```
