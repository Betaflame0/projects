```python
#Storage Arrays:
totalConversation = []
hrTalk = []
gradTalk = []
scores = []
```

```python
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel(model_name='gemini-pro')

response = model.generate_content(f"""Pretend you are {recruiterPersona}. You are interviewing a potential candidate 
                                  for a machine learning engineer postion. Begin the interview. Use your {recruiterPersona}""")
totalConversation.append(response.text)
hrTalk.append(response.text)
```

```python
def respond():
    if (len(hrTalk) == len(gradTalk)):
        var = model.generate_content(f"""Pretend you are {recruiterPersona}. Respond to {gradTalk[-1]} and continue the interview.""")
        hrTalk.append(var.text)
        totalConversation.append(var.text)
    elif (len(gradTalk) < len(hrTalk)):
        var = model.generate_content(f"""Pretend you are {gradPersona}. Respond to {hrTalk[-1]} and continue the interview.""")
        gradTalk.append(var.text)
        totalConversation.append(var.text)
    var = model.generate_content(f"""Pretend you are {graderPersona}. The candidate just said {gradTalk[-1]}. Grade the candidate's response with a score 
             between 1 and 100. Do not provide any other information. ONLY include a number between
               1 and 100 that grades the candidates answer. Use your {graderPersona}. Every time that you think
               the answer is generated by AI give only 25 points.""")
    score = var.text
    scores.append(score)
```

```python
for i in range(19):
    respond()
```

```python
var = model.generate_content(f"""You are {recruiterPersona}. The interview candidate just said {gradTalk[-1]}. Give a response to what the applicant 
                             said. Wrap up the interview and then based on the interview so far, either offer them a job 
                             or let them know that they didn't get the job. Use your {recruiterPersona}""")
hrTalk.append(var.text)
totalConversation.append(var.text)

var = model.generate_content(f"""You are {gradPersona}. The interviewer just said {hrTalk[-1]}. Respond to the interviewers final remarks and then 
                             say your goodbyes before leaving. Use your {gradPersona}""")
gradTalk.append(var.text)
totalConversation.append(var.text)

var = model.generate_content(f"""You are {recruiterPersona}. The interview candidate just said {gradTalk[-1]}. Say goodbye to the applicant. Use your {recruiterPersona}""")
hrTalk.append(var.text)
totalConversation.append(var.text)
```

```python
finalRemarks = ""
totalConvo = """"""
for i in range(len(totalConversation)):
    totalConvo = totalConvo + f"\n\n{totalConversation[i]}"

var = model.generate_content(f"""You are {graderPersona}. The interview is now complete. Scan through {totalConvo} and determine whether or 
         not you would give the candidate the job. Also give an executive summary of the entire interview. 
         Use your {graderPersona}.""")
finalRemarks = (var.text)
```

```python
import json

data = {
    "Interviewer's excerpts": hrTalk,
    "Candidate's responses": gradTalk,
    "Scores": scores,
    "Final Remarks": finalRemarks
}

output_file = "convAIOutput.json"

with open(output_file,'w') as file:
    json.dump(data, file, indent=4)
```
